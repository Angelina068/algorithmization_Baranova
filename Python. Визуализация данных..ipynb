```python
import matplotlib.pyplot as plt

# Данные
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]

# Создание столбчатой диаграммы
plt.bar(categories, values, color='skyblue')
plt.title('Столбчатая диаграмма')
plt.xlabel('Категории')
plt.ylabel('Значения')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Данные
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
colors = ['gold', 'lightcoral', 'lightskyblue', 'yellowgreen']

# Создание круговой диаграммы
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Круговая диаграмма')
plt.axis('equal')  # Чтобы круг был кругом
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Данные
```
```python
data = np.random.randn(1000)

# Создание гистограммы
plt.hist(data, bins=30, color='green', alpha=0.7)
plt.title('Гистограмма')
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.grid(True)
plt.show()
```

•	 subplot() — функция для задания местоположения поля с графиком. Существует несколько способов задания областей для вывода графиков.
```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Создаем фигуру и оси
fig, ax = plt.subplots()
x = np.linspace(0, 2 * np.pi, 100)
line, = ax.plot(x, np.sin(x))

# Устанавливаем пределы осей
ax.set_xlim(0, 2 * np.pi)
ax.set_ylim(-1.5, 1.5)

# Функция обновления для анимации
def update(frame):
    line.set_ydata(np.sin(x + frame / 10))  # Изменяем данные Y
    return line,

# Создаем анимацию
ani = FuncAnimation(fig, update, frames=100, interval=50)

# Показываем график
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Создаем фигуру и оси
fig, ax = plt.subplots(figsize=(6, 6))

# Настройки для кругов
num_circles = 10
colors = plt.cm.viridis(np.linspace(0, 1, num_circles))

# Рисуем круги
for i in range(num_circles):
    circle = plt.Circle((0.6, 0.6), 0.4 * (num_circles - i) / num_circles, color=colors[i], fill=True)
    ax.add_artist(circle)

# Устанавливаем пределы и убираем оси
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')
ax.axis('off')

# Показываем изображение
plt.title("")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Создаем фигуру и оси
fig, ax = plt.subplots(figsize=(8, 8))

# Настройки для полос
num_stripes = 30
stripe_width = 1.0 / num_stripes
x = np.linspace(0, 1, 100)

# Рисуем полосы
for i in range(num_stripes):
    # Чередуем цвета
    color = 'black' if i % 2 == 0 else 'white'
    ax.fill_between(x, i * stripe_width, (i + 1) * stripe_width, color=color)

# Устанавливаем пределы и убираем оси
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')
ax.axis('off')

# Добавляем текст для пояснения
plt.title("")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Создаем фигуру и оси
fig, ax = plt.subplots(figsize=(6, 6))

# Настройки для квадратов
num_squares = 10
size = 1 / num_squares
colors = ['black', 'white']

# Рисуем квадраты
for i in range(num_squares):
    for j in range(num_squares):
        color = colors[(i + j) % 2]  # Чередуем цвета
        square = plt.Rectangle((j * size, i * size), size, size, color=color)
        ax.add_artist(square)

# Устанавливаем пределы и убираем оси
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')
ax.axis('off')

# Добавляем текст для пояснения
plt.title("")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Создаем фигуру
fig, ax = plt.subplots(figsize=(6, 6))

# Кот - голова
head = plt.Circle((0.5, 0.5), 0.3, color='gray', ec='black')

# Кот - уши
left_ear = plt.Polygon([[0.25, 0.75], [0.35, 0.9], [0.4, 0.75]], color='gray', ec='black')
right_ear = plt.Polygon([[0.75, 0.75], [0.65, 0.9], [0.6, 0.75]], color='gray', ec='black')

# Кот - глаза
left_eye = plt.Circle((0.4, 0.5), 0.05, color='white', ec='black')
right_eye = plt.Circle((0.6, 0.5), 0.05, color='white', ec='black')
left_pupil = plt.Circle((0.4, 0.5), 0.02, color='black')
right_pupil = plt.Circle((0.6, 0.5), 0.02, color='black')

# Кот - нос
nose = plt.Circle((0.5, 0.45), 0.03, color='pink')

# Кот - рот
x = np.array([0.5, 0.48, 0.52])
y = np.array([0.4, 0.42, 0.42])
plt.plot(x, y, color='black')

# Добавляем элементы на график
ax.add_artist(head)
ax.add_artist(left_ear)
ax.add_artist(right_ear)
ax.add_artist(left_eye)
ax.add_artist(right_eye)
ax.add_artist(left_pupil)
ax.add_artist(right_pupil)
ax.add_artist(nose)

# Устанавливаем пределы и убираем оси
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')
ax.axis('off')

# Показываем изображение
plt.title("")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Определяем размеры изображения
width, height = 400, 300

# Создаем массив для хранения цветов
gradient = np.zeros((height, width, 3))

# Заполняем массив градиентом
for x in range(width):
    for y in range(height):
        # Создаем градиент от красного к синему
        gradient[y, x, 0] = x / width   # Красный
        gradient[y, x, 1] = 0            # Зеленый
        gradient[y, x, 2] = (height - y) / height  # Синий

# Создаем график
plt.imshow(gradient, aspect='auto')
plt.axis('off')  # Убираем оси
plt.title('')
plt.show()
```
